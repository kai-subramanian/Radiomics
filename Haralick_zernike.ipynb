```python
import mahotas
from pylab import gray,imshow,show
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageEnhance
```

```python
#we'll look at haralick moments first
img = mahotas.imread('/home/kailash/DIP Project/archive/AbdomenCT/000017.jpeg')
gaussian = mahotas.gaussian_filter(img,15)
gaussian = (gaussian>gaussian.mean())
label,n = mahotas.label(gaussian)
```

```python
print("Labelled image")
imshow(label)
show()
h_feat = mahotas.features.haralick(label)
print("Haralick Features")
imshow(h_feat)
show()
```

```python
#now, zernike moments
img = Image.open("/home/kailash/DIP Project/archive/AbdomenCT/000000.jpeg")
#sharpen the image
enhancer = ImageEnhance.Sharpness(img)
factor=1.1
img = enhancer.enhance(factor)
img.show()
img.save('/home/kailash/DIP Project/archive/AbdomenCT/sharpened-image.png');
img = mahotas.imread('/home/kailash/DIP Project/archive/AbdomenCT/sharpened-image.png')
print("Image:")
imshow(img)
show()
deg=10
rad=10
val=mahotas.features.zernike_moments(img,rad,deg)
print(val)
```
